```python
# using mobile app https://play.google.com/store/apps/details?id=de.lorenz_fenster.sensorstreamgps&hl=en
import pandas as pd
import socket, traceback

host = ''
port = 5555

s = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
s.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
s.setsockopt(socket.SOL_SOCKET, socket.SO_BROADCAST, 1)
s.bind((host, port))

data = pd.DataFrame()
i=0
while i<100:
    try:
        message, address = s.recvfrom(8192)
        # print(message)
        print(str(message).split())
        data = data.append(pd.DataFrame(message.split()).T)
        i += 1
    except (KeyboardInterrupt, SystemExit):
        raise
    except:
        traceback.print_exc()
```
