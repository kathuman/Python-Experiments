```python
# Importing the dataset
dataset=pd.read_excel('Data_Gokul.xlsx', sheet_name='Data Sheet')

# Creating the independent variable Set
X=pd.DataFrame(dataset.iloc[46:268,1:4])

# Vreating the Dependet variable set
y = pd.DataFrame(dataset.iloc[46:268,4 ])
```

```python
# We graph the dependent variable to se its evolution over time
y.plot(figsize=(16,10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.title('Dependent Variable')
plt.xlabel('Week number')
plt.ylabel('Y-Value: Outgoing Pallets')
plt.show()
```

```python
#We take a look at the relationship between the variables in X
sns.pairplot(X, corner=True)
plt.show()
```

```python
# Analyzing the shape of the X Vector, to determine how to split it. It cannot be split randomly as it is a data series.
X.shape
```

```python
# New form of splitting, by week number
TEST_SIZE = 55 # Number of days used for the testing

X_test = X.tail(TEST_SIZE)
print(X_test.shape)
print(X_test.tail(TEST_SIZE).head())
print(X_test.tail(TEST_SIZE).tail(10))
```

```python
# Training size is the difference
X_train = X.head(X.shape[0]-TEST_SIZE)
print(X_train.shape)
print(X_train.head())
print(X_train.tail())
```

```python
# we do the same with the Y values
y_test = y.tail(TEST_SIZE)
y_train = y.head(X.shape[0]-TEST_SIZE)
print(y_test.shape)
print(y_train.shape)
print(y_test)
```

```python
plt.plot(y_train, label = 'y_train')
plt.plot(y_test, label = 'y_test')
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.title('y-values')
plt.xlabel('Week number')
plt.ylabel('Outgoing Pallets')
plt.rcParams["figure.figsize"] = (20,10)
plt.legend()
plt.show()
```

# Using Multiple Regression using full dataset

```python
#fitting_multiple Regression model 
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
print("Coefficients: ",regressor.coef_)
print("Intercept: ",regressor.intercept_)
```

```python
# Inorder to get some statistics of the result
import statsmodels.api as sm
results = sm.OLS(y_train,X_train).fit()
results.summary()
```

```python
#predict_test set
y_pred=pd.DataFrame(regressor.predict(X_test))
```

```python
y_pred['id']=range(213,268)
y_pred.set_index('id', inplace = True, append = False, drop = True)
y_pred.columns = ['y_pred']
print(y_pred.head(3))
print(y_pred.tail(3))
```

```python
y_comparison = y_test
y_comparison['y_pred'] = y_pred
y_comparison.columns = ['y_test','y_pred']
print(y_comparison.head())
print(y_comparison.tail(10))

y_comparison.plot(kind='bar',figsize=(16,10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()

print(X_test.tail(10))
```
