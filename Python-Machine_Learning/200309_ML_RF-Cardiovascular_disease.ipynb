```python
# Remove or drop the years column
df = df.drop('years', axis=1)
```

```python
df = df.drop('id', axis=1)
```

```python
#Split the data into feature data and target data
X = df.iloc[:,:-1].values
Y = df.iloc[:,-1].values
```

```python
#Split the data again into 75% training and 25% testing
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.25, random_state=1)
```

```python
# We now do feature scaling (scale values in data to be values between 0 and 1)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Use Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=1)
forest.fit(X_train,Y_train)
```

```python
# Test model accuracy on the training data set
model = forest
model.score(X_train, Y_train)
```

```python
#test model accuracy in the test data set
from sklearn.metrics import confusion_matrix
Y_predictions = model.predict(X_test)
cm = confusion_matrix(Y_test, Y_predictions)

TN = cm[0][0]
TP = cm[1][1]
FN = cm[1][0]
FP = cm[0][1]

# Print the confuision Matrix
print(cm)

#Print the model accuracy
print('Model Accuracy = {}'.format((TN+TP)/(TN+TP+FN+FP)))
```

```python
### ROC and AUC Curves ###
# https://towardsdatascience.com/understanding-auc-roc-curve-68b2303cc9c5
# Generate a ROC Curve
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

#No Skill Predictor
ns_probs = [0 for _ in range(len(Y_test))]

# calculate scores
ns_auc = roc_auc_score(Y_test, ns_probs)
lr_auc = roc_auc_score(Y_test, Y_predictions)

# summarize scores
print('No Skill: ROC AUC=%.3f' % (ns_auc))
print('Random forest predictor: ROC AUC=%.3f' % (lr_auc))

# calculate roc curves
ns_fpr, ns_tpr, _ = roc_curve(Y_test, ns_probs)
lr_fpr, lr_tpr, _ = roc_curve(Y_test, Y_predictions)
# plot the roc curve for the model
plt.plot(ns_fpr, ns_tpr, linestyle='--', label='No Skill')
plt.plot(lr_fpr, lr_tpr, marker='.', label='Random Forest')
# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# show the legend
plt.legend()
# show the plot
plt.show()
```
