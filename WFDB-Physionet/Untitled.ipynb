```python
# import the WFDB package
import wfdb
```

```python
# load a record using the 'rdrecord' function
# LOADS ALL CHANNELS
#https://wfdb.readthedocs.io/en/latest/wfdb.html?highlight=wfdb.rdrecord#wfdb.rdrecord
record = wfdb.rdrecord('sample-data2/1')
```

```python
# plot the record to screen
# https://wfdb.readthedocs.io/en/latest/plot.html
wfdb.plot_wfdb(record=record, title='Example signals',figsize=(10,40))
```

```python
record = wfdb.rdrecord('sample-data2/1', channels=[0,1])
```

```python
wfdb.plot_wfdb(record,figsize=(25,10), sig_style=['classic'])
```
